```python
from pymodulon.core import IcaData
from pymodulon.plotting import *
from pymodulon.io import load_json_model
import pandas as pd

from os import path
import matplotlib.patches as mpatches

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['font.sans-serif']='Arial'
```

### CueR iMod Activity vs. brlR gene expression

```python
data=pd.read_csv("../data/processed_data/TCS_blrR_input.csv", sep="\t")
data.head()
```

```python
sns.set_style('ticks')
fig, ax = plt.subplots()
fig.set_size_inches(20, 5)
sns.lineplot(data=data, x="iMod_name", y="PCC")
sns.despine()
ax.tick_params(axis='x', rotation=90)
fig.savefig("../../../UCSD/Pseudo_Precise/Figures/paeru_precise02/TCS_brlR.pdf", transparent=True, bbox_inches = 'tight')
```
