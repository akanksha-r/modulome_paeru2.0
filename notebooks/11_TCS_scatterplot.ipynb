```python
from pymodulon.core import IcaData
from pymodulon.plotting import *
from pymodulon.io import load_json_model
import pandas as pd

from os import path
import matplotlib.patches as mpatches

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['font.sans-serif']='Arial'
```

```python
ica_data = load_json_model('../../modulome_paeru2.0/paeru_116_iModulon.json')
```

### CzcR iModulon vs. czcR gene

```python
scatterplot(ica_data.X.T['PA2523'],ica_data.A.T['CzcR'],
            fit_metric= 'pearson',fit_line='True',
           xlabel='czcR expression', ylabel='CzcR iModulon activity')

# plt.savefig("../../../UCSD/Pseudo_Precise/Figures/paeru_precise02/scatter_plot_CzcR_iMod_czcR_gene.pdf", transparent=True, bbox_inches = 'tight')
```
